```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import AdaBoostClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
```

```python
raw_data=load_breast_cancer()
df=pd.DataFrame(raw_data.data,columns=raw_data.feature_names)
y = pd.Categorical.from_codes(raw_data.target, raw_data.target_names)
```

```python
df.head()
```

```python
y
```

```python
encoder=LabelEncoder()
bimary_y=pd.Series(encoder.fit_transform(y))
```

```python
X_train,X_test,y_train,y_test=train_test_split(df,bimary_y,test_size=0.2,random_state=42)
```

```python
clf_1=AdaBoostClassifier(
    DecisionTreeClassifier(max_depth=1),
    n_estimators=200
)
clf_1.fit(X_train,y_train)
```

```python
y_pred=clf_1.predict(X_test)
```

```python
confusion_matrix(y_test,y_pred)
```

```python
clf_2=RandomForestClassifier(n_estimators=200)
clf_2.fit(X_train,y_train)
```

```python
y_pred_2=clf_2.predict(X_test)
confusion_matrix(y_test,y_pred_2)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
clf= GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,max_depth=1,random_state=0)
clf.fit(X_train,y_train)
```

```python
y_pred3=clf.predict(X_test)
y_pred3_prob=clf.predict_proba(X_test)
confusion_matrix(y_test,y_pred3)
```
